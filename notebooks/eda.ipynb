```python
# üìå 1. Initial Setting
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Make visualization immediately
%matplotlib inline

# üìå 2. Read data
df = pd.read_csv("../data/melb_housing.csv")  # Í≤ΩÎ°ú ÌôïÏù∏!
print(f"Dataset shape: {df.shape}")
df.head()


```

```python
# Visualize number of missing values
plt.figure(figsize=(12,6))
df.isnull().sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Missing values per column")
plt.show()
```

```python
# Price Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.show()

# Building Area Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['BuildingArea'].dropna(), bins=50, kde=True)
plt.title('Building Area Distribution')
plt.xlabel('Area')
plt.show()
```

```python
# Extract numerical variables
num_df = df.select_dtypes(include=[np.number])

# Correlation matrix
corr = num_df.corr()

# Visualize correlation matrix
plt.figure(figsize=(12,10))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Feature Correlation Matrix")
plt.show()
```
